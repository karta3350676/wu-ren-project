```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
sns.set_style("darkgrid")

import warnings
warnings.filterwarnings("ignore")
dfs = dict()
dfs["transaction_data"] = pd.read_csv("transaction_data.csv")
dfs["product_data"] = pd.read_csv("product.csv")
dfs["transaction_data"].head()
```

```python
dfs["product_data_onlyID"] = dfs["product_data"]["SUB_COMMODITY_DESC"]
dfs["product_data_onlyID"] .head(100)
```

```python
dfs["transaction_data_2"]=pd.merge(dfs["product_data"],dfs["transaction_data"], on='PRODUCT_ID')
dfs["transaction_data_2"]
```

```python
# df1 = dfs["transaction_data_2"]
# indexNames = df1["PRODUCT_ID"] == '6534178' 
# dfs["product_data_drop"]=dfs["transaction_data_2"].drop(indexNames ,axis=0)
# dfs["product_data_drop"]
```

```python
sales_per_PRO = dfs["transaction_data_2"].groupby("PRODUCT_ID").sum()["QUANTITY"].sort_values(ascending=False)
sales_per_PRO = pd.DataFrame(list(zip(sales_per_PRO.index,sales_per_PRO)),columns=["PRODUCT_ID","Total_Q"])
fig_store = plt.figure(figsize=(14,8))
sns.barplot(y="PRODUCT_ID",x="Total_Q",data = sales_per_PRO[1:30],order=sales_per_PRO[1:30]["PRODUCT_ID"],orient="h")
plt.title('Top 20 PRODUCT based on Sale Amount', fontsize=17)
plt.xlabel('Total_Q', fontsize=14)
plt.ylabel('PRODUCT_ID', fontsize=14)
plt.show()
```

```python
sales_per_PRO = dfs["transaction_data_2"].groupby("SUB_COMMODITY_DESC").sum()["QUANTITY"].sort_values(ascending=False)
sales_per_PRO = pd.DataFrame(list(zip(sales_per_PRO.index,sales_per_PRO)),columns=["SUB_COMMODITY_DESC","Total_Q"])
fig_store = plt.figure(figsize=(14,8))
sns.barplot(y="SUB_COMMODITY_DESC",x="Total_Q",data = sales_per_PRO[1:30],order=sales_per_PRO[1:30]["SUB_COMMODITY_DESC"],orient="h")
plt.title('Top 20 PRODUCT based on Sale Amount', fontsize=17)
plt.xlabel('Total_Q', fontsize=14)
plt.ylabel('SUB_COMMODITY_DESC', fontsize=14)
plt.show()
```
